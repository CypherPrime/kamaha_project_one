# Now working on handling the missing values

````python
crashes.head()
````

````python
# print missing values for crashes
print(crashes.isnull().sum())
````

````python
crashes_cols = crashes.columns

for cols in  crashes_cols:
    print("{} : {} : {}".format(cols, crashes[cols].isnull().sum(), crashes[cols].dtype))
````

````python
for cols in  crashes_cols:
    if crashes[cols].dtype == 'object':
        crashes[cols].fillna('Unknown', inplace=True) # if we were to build a model this would be used 
    else:
        crashes[cols].fillna(crashes[cols].mean(), inplace=True)
````

# Cleaning People.csv

````python
people_cols = people.columns

for cols in  people_cols:
    if people[cols].dtype == 'object':
        people[cols].fillna('Unknown', inplace=True) # if we were to build a model this would be used 
    else:
        people[cols].fillna(people[cols].mean(), inplace=True)
````

# Encode Spatial Data (GeoHash Example)

````python
!pip install python-geohash
````

````python
from geopy.geocoders import Nominatim
import geohash
````

````python
# Example of encoding latitude and longitude into GeoHash
# this adds a column on our dataframe called GEOHASH containing th geohash values for each row
def encode_geohash(row):
    return geohash.encode(row['LATITUDE'], row['LONGITUDE'])

# Apply GeoHash encoding to crashes DataFrame
crashes['GEOHASH'] = crashes.apply(encode_geohash, axis=1)
````

````python
import pyodbc
import mysql.connector
````

````python
# Connect to SQL Server :: this is if we are using sql server
conn = pyodbc.connect('DRIVER={SQL Server};'
                      'SERVER=your_server;'
                      'DATABASE=your_database;'
                      'UID=your_username;'
                      'PWD=your_password;')
````

````python
# Connect to MySQL : if using xampp
conn = mysql.connector.connect(
    host='localhost',
    user='your_username',  # Replace with your MySQL username
    password='your_password',  # Replace with your MySQL password
    database='your_database'  # Replace with your database name
)
````

````python
cursor = conn.cursor()

cursor.execute('''
CREATE TABLE Crashes (
    RD_NO NVARCHAR(50),
    CRASH_DATE DATETIME,
    POSTED_SPEED_LIMIT INT,
    GEOFASH NVARCHAR(12),
)
''')

cursor.execute('''
CREATE TABLE People (
    PERSON_ID NVARCHAR(50),
    RD_NO NVARCHAR(50),
    CITY NVARCHAR(50),
    SEX NVARCHAR(10),
)
''')

cursor.execute('''
CREATE TABLE Vehicles (
    VEHICLE_ID NVARCHAR(50),
    RD_NO NVARCHAR(50),
    MAKE NVARCHAR(50),
)
''')

conn.commit()
````

````python
# Uploading our DataFrames to SQL Server
crashes.to_sql('Crashes', conn, if_exists='append', index=False)
people.to_sql('People', conn, if_exists='append', index=False)
vehicles.to_sql('Vehicles', conn, if_exists='append', index=False)
````

````python
# Now closing the connection
conn.close()
````

````python
# Alternatively, 
# Function to upload a DataFrame to MySQL
def upload_to_mysql(df, table_name):
    for row in df.itertuples(index=False):
        cursor.execute(f'''
            INSERT INTO {table_name} VALUES (%s, %s, %s, %s)
        ''', row)

# Upload DataFrames to MySQL
upload_to_mysql(crashes, 'Crashes')
upload_to_mysql(people, 'People')
upload_to_mysql(vehicles, 'Vehicles')

# Commit the transaction
conn.commit()

# Close the connection
cursor.close()
conn.close()
````
